```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result.csv')
df
```

```python
# Category 단위

sc = plt.scatter(df['Uniformity'], df['Alignment (Category)'], s=70, c=df['Downstream Task (Category)'], cmap='Greens')
plt.xlim([-3, 0])
plt.xlabel("Uniformity")
plt.ylabel("Alignment (Category)")

plt.colorbar(sc).ax.set_ylabel('Classification Accuracy', rotation=270, labelpad=15)

for i, txt in enumerate(df['Model (Category)']):
    plt.annotate(txt, (df['Uniformity'][i], df['Alignment (Category)'][i]))
```

```python
# Class 단위

sc = plt.scatter(df['Uniformity'], df['Alignment (Class)'], s=70, c=df['Downstream Task (Class)'], cmap='Greens')
plt.xlim([-3, 0])
plt.xlabel("Uniformity")
plt.ylabel("Alignment (Class)")

plt.colorbar(sc).ax.set_ylabel('Classification Accuracy', rotation=270, labelpad=15)

for i, txt in enumerate(df['Model (Class)']):
    plt.annotate(txt, (df['Uniformity'][i], df['Alignment (Class)'][i]))
```
